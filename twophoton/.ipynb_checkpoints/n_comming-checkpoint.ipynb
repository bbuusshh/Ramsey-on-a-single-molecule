```python
from matplotlib import pyplot as plt
import numpy as np
import os
from tools import tools as tl
DUMP_TYPE = np.dtype([("time_from_trigger", np.dtype("<u4")),
        ("channel", np.dtype("<i4")),
        ("time", np.dtype("<u8")),])

# os.listdir("C:\Data")
os.chdir("C:\Data")
```

```python
data = np.fromfile("somenottoobad_3.dump",dtype = DUMP_TYPE, count=int(1e7))

trig = tl.get_trig_len(data)
data = tl.shift_channels(data, -70000, 5)
data = tl.shift_channels(data, 50_000, 8)
data = data[np.argsort(data['time'])]
data0 = tl.assign_trigs_to_clicks(data, channels=[1,8], trig_ch=5)
data0 = data0[np.argsort(data0['time'])]
# data0 = tl.assign_trigs_to_clicks(data, channels=[1,8], trig_ch=5)

data = data0[data0['channel']!=5]

y, x = np.histogram(data['time_from_trig'][(data['channel']==8)], bins=1000, range=(0, trig))
plt.plot(x[:-1], y)
y, x = np.histogram(data['time_from_trig'][(data['channel']==1)], bins=1000, range=(0, trig))
plt.plot(x[:-1], y)
```
